```python
import numpy as np

# Transition matrix
P = np.array([
    [1.0, 0.0, 0.0],
    [0.4, 0.4, 0.2],
    [0.0, 0.6, 0.4]
])

# Initial beliefs as row vectors
bel_0_1 = np.array([1.0, 0.0, 0.0])  # sunny
bel_0_2 = np.array([0.0, 1.0, 0.0])  # cloudy
bel_0_3 = np.array([0.0, 0.0, 1.0])  # rainy

# Number of steps
t = 100

# Compute P^t
P_t = np.linalg.matrix_power(P, t)

# Compute beliefs at step t
bel_t_1 = bel_0_1 @ P_t
bel_t_2 = bel_0_2 @ P_t
bel_t_3 = bel_0_3 @ P_t


print("Belief at step" ,t, "starting from sunny:", bel_t_1)
print("Belief at step" ,t, "starting from cloudy:", bel_t_2)
print("Belief at step" ,t, "starting from rainy:", bel_t_3)
```

```python
import numpy as np

A = np.array([
    [-0.2, 0.4, 0.2],
    [0.2, -0.6, 0.6],
    [1, 1, 1]
])

b = np.array([0, 0, 1])

sol1 = np.linalg.lstsq(A, b, rcond=None)[0]
print(sol1)

sol = np.linalg.solve(A,b)
print(sol)
```

```python
import sympy as sp

# Unknowns
x, y, z = sp.symbols('x y z', real=True)

# Independent equations from wP = w
eq1 = sp.Eq(0.8*x + 0.4*y + 0.2*z, x)
eq2 = sp.Eq(0.2*x + 0.4*y + 0.6*z, y)

# Normalization w = x + y + z = 1
eq3 = sp.Eq(x + y + z, 1)

# Solve
sol = sp.solve((eq1, eq2, eq3), (x, y, z), set=True)
sol
```
